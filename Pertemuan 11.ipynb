## Regression Models

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
x = np.array([5, 15, 25, 35, 45, 55])##.reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
print (x)
print (y)
```

```python
x.reshape(-1, 1)
```

```python
plt.plot(x, y)
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Define Model
model=LinearRegression()
```

```python
model
```

```python
# Train Linear Regression Model
model.fit(x.reshape(-1, 1), y)
```

```python
model.coef_, model.intercept_
```

```python
lr_score=model.score(x.reshape(-1, 1), y)
print('R-Squared:', lr_score)
```

```python
y_hat=model.predict(x.reshape(-1, 1))
print('Input:', x)
print('Target:', y)
print('Prediction:', y_hat)
```

```python
plt.plot(x, y, label='Target')
plt.plot(x, y_hat, label='Prediction', color='red')
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
print(model.coef_, model.intercept_)
```

```python
# Input 0
model.predict([[1]])
```

## Multiple Linear Regression

```python
x=np.array([
    [0, 1],
    [5, 1],
    [15, 2],
    [25, 5],
    [35, 11],
    [45, 15],
    [55, 34],
    [60, 35],
])
y=np.array([4, 5, 20, 14, 32, 22, 38, 43])
print(x)
print(y)
```

```python
#Define Model
model = LinearRegression()

#Fit Model
model.fit(x, y)

#Evaluate Model
r2=model.score(x, y)
print('R-Squared:', r2)

#Predict
y_hat=model.predict(x)

#Plot
#Plot Fitur 1
plt.plot(x[:, 0], y, label='Target')
plt.plot(x[:, 0], y_hat, label='Prediction', color='red')
plt.scatter(x[:, 0], y)
plt.xlabel('X1')
plt.ylabel('Y')
plt.legend()
plt.show()
#Plot Fitur 2
plt.plot(x[:, 1], y, label='Target')
plt.plot(x[:, 1], y_hat, label='Prediction', color='red')
plt.scatter(x[:, 1], y)
plt.xlabel('X2')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
print('Model Weights:', model.coef_)
print('Model Bias:', model.intercept_)
```

```python
model.predict([[0, 1]]), model.predict([[1, 0]])
```

## Polynomial Regression

```python
x=np.array([5, 15, 25, 35, 45, 55])
y=np.array([15, 11, 2, 8, 25, 32])
print(x)
print(y)
```

```python
plt.plot(x, y)
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
polynom=PolynomialFeatures(degree=2, include_bias=False)
polynom
```

```python
x
```

```python
polynom.fit(x.reshape(-1, 1))
```

```python
poly_X=polynom.transform(x.reshape(-1, 1))
```

```python
print(x.reshape(-1, 1))
print(poly_X)
```

```python
model=LinearRegression()
model.fit(poly_X, y)
r2=model.score(poly_X, y)
print('R2 Score:', r2)
print('Coef:', model.coef_)
print('intercept:', model.intercept_)

y_hat=model.predict(poly_X)

plt.plot(x, y, label='Target')
plt.plot(x, y_hat, label='Prediction', color='Red')
plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

## Linear Regression with `Statsmodels`

```python
x=np.array([
    [0, 1],
    [5, 1],
    [15, 2],
    [25, 5],
    [35, 11],
    [45, 15],
    [55, 34],
    [60, 35],
])
y=np.array([4, 5, 20, 14, 32, 22, 38, 43])
print(x)
print(y)
```

```python
x=sm.add_constant(x)
print(x)
print(y)
```

```python
model=sm.OLS(y, x)
```

```python
model
```

```python
results=model.fit()
```

```python
print(results.summary())
```
